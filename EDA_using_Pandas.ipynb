**Data Preparation**

```python
#Checking Missing Values
df.isnull().sum()
```

```python
# Checking duplicates
duplicates = df[df.duplicated()]

# If there are duplicates, duplicates will contain those rows
if not duplicates.empty:
    total_duplicates = duplicates.shape[0]
    print("Total number of duplicate rows:", total_duplicates)
else:
    print("No duplicate rows found.")
```

**Handling Missing Values**

```python

df['director'] = df['director'].fillna('Unknown').astype('category')
df.head(5)
```

```python
#Dropping null values
df.dropna(inplace = True)
```

```python
#Checking Missing Values
df.isnull().sum()
```

```python
df.head(5)
```

```python
df = df.rename(columns={'date_added':'Released'})
df.head(5)
```

**Checking Correlation**

```python
corr_matrix = df.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(5, 3))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

**Correlation btw non-numerals parameters**

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['director'], df['rating'])

# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)
```

```python
print("Chi-Square Statistic:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies Table:")
print(expected)
```

```python
#splitting
df['added_month'] = df['date_added'].apply(lambda x:x.split(',')[0].split()[0])
```

```python
df.head(10)
```

**Converting values in numerals**

```python
df['is_movie'] = df['type'].apply(lambda x: 1 if x == 'Movie' else 0)
```

```python
df.head(5)
```

```python
# Drop the specified columns
df.drop(['type'], axis=1, inplace=True)
```

```python
corr_matrix = df.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(5, 3))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
temp_df = df[df['country'].isin(df['country'].value_counts()[:5].index)]
sns.countplot(data=temp_df,x='country',hue='type')
```

```python
top_countries = df['country'].value_counts()[:5].index
temp_df = df[df['country'].isin(top_countries)]

# Create a new '5-year interval' column to aggregate by 5-year intervals
temp_df['5-year interval'] = ((temp_df['release_year'] - 1) // 5) * 5

# Create a countplot with data aggregated by 5-year intervals
plt.figure(figsize=(12, 6))
sns.countplot(data=temp_df, x='5-year interval', hue='country')
plt.title("Number of Netflix Titles Released by 5-Year Intervals and Country (Top 5)")
plt.xlabel("5-Year Interval")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
